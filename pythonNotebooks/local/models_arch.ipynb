```python
from keras.models import model_from_json
```

```python
MODELS_CNT = 9

cnn_models = []

for model_indice in range(1, MODELS_CNT + 1):
    with open('../../trained_models/final/arch_{}.json'.format(model_indice)) as fl:
        model = model_from_json(fl.read())
    model.load_weights('../trained_models/final/weights_{}.h5'.format(model_indice))
    cnn_models.append(model)
```

```python
model = cnn_models[0]
```

```python
layers_names = []
for model in cnn_models:
    modelN = []
    for layer in model.layers:
        modelN.append(layer.name)
    layers_names.append(modelN)
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(layers_names).transpose()
df.fillna(value='', inplace=True)
df.columns = range(1, 10)
df.index = range(1, 43)
df
```

```python
df.replace(value='', regex='\_[0-9]{1,2}', inplace=True)
df
```

```python
mapa = {
    'conv2d' : 'Conv2D',
    'batch_normalization' : 'BatchNorm',
    'max_pooling2d' : 'MaxPool',
    'dropout' : 'Dropout',
    'flatten' : 'Flatten',
    'dense' : 'FC',
    'average_pooling2d' : 'AvgPool'   ,
    'activation' : ''
}
```

```python
for key, value in mapa.items():
    df.replace(key, value, inplace=True)
df
```

```python
df.to_csv('../eniac2018/source/archs.csv', index=False)
```
